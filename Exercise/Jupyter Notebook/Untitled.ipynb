```micropython
import time
from lsm6dsox import LSM6DSOX
from machine import Pin, I2C

# Initialize the LSM6DSOX sensor
lsm = LSM6DSOX(I2C(0, scl=Pin(13), sda=Pin(12)))

# Function to calculate the delta value
def calculate_delta(prev, curr):
    return [curr[i] - prev[i] for i in range(3)]

# Initialize previous data with the first reading
prev_accel_data = lsm.accel()
prev_gyro_data = lsm.gyro()

# Run the loop 100 times
for _ in range(200):
    # Get current accelerometer and gyroscope data
    curr_accel_data = lsm.accel()
    curr_gyro_data = lsm.gyro()
    
    # Calculate delta values for accelerometer and gyroscope
    delta_accel = calculate_delta(prev_accel_data, curr_accel_data)
    delta_gyro = calculate_delta(prev_gyro_data, curr_gyro_data)
    
    # Print the formatted data and delta values in one row
    print('{:>8.3f},{:>8.3f},{:>8.3f}, {:>8.3f},{:>8.3f},{:>8.3f}, {:>8.3f},{:>8.3f},{:>8.3f},{:>8.3f},{:>8.3f},{:>8.3f},2'.format(
          *curr_accel_data, *delta_accel, *curr_gyro_data, *delta_gyro))
    
    # Update previous data
    prev_accel_data = curr_accel_data
    prev_gyro_data = curr_gyro_data
    
    # Wait for 100 ms
    time.sleep_ms(100)
```
